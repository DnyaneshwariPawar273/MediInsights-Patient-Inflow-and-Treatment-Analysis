```python
import pandas as pd
df=pd.read_csv("Hospital_Patient_Analytics.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#converting date columns to datetime
df['Admission_Date'] = pd.to_datetime(df['Admission_Date'])
df['Discharge_Date'] = pd.to_datetime(df['Discharge_Date'])
df
```

```python
# setting addmission date of outpaitent equals to its discharde date
df.loc[
    df['Patient_Type'] == 'Outpatient',
    'Discharge_Date'] = df.loc[
    df['Patient_Type'] == 'Outpatient',
    'Admission_Date']
df
```

```python
# finding minimum inpatient cost
min_inpatient_cost = df.loc[
    df['Patient_Type'] == 'Inpatient',
    'Treatment_Cost'
].min()
min_inpatient_cost
```

```python
# managing invalid outpatient cost ( outpatient_cost<inpatient_cost)
outpatient_cost = round(min_inpatient_cost * 0.7) # asuming that outpatient cost is 70% of minimum inpatient cost
df.loc[
    (df['Patient_Type'] == 'Outpatient') &
    (df['Treatment_Cost'] >= min_inpatient_cost),
    'Treatment_Cost'
] = outpatient_cost

df

```

```python
#checking(outpatient_cost<inpatient_cost)
df.groupby('Patient_Type')['Treatment_Cost'].describe()
```

```python
# cheking addmission and discharge dates of outpaitent
df[df['Patient_Type'] == 'Outpatient'][['Admission_Date','Discharge_Date']]
```

```python
# extracting year month and day from date 
df['Admission_Month'] = df['Admission_Date'].dt.month_name()
df['Discharge_Month'] = df['Discharge_Date'].dt.month_name()
df['Admissin_Year'] = df['Admission_Date'].dt.year
df['Discharge_Year'] = df['Discharge_Date'].dt.year
df['Admission_Weekday'] = df['Admission_Date'].dt.day_name()
df['Discharge_Weekday'] = df['Discharge_Date'].dt.day_name()
df
```

```python
df.duplicated()
```

```python
#creating two different columns for inpatient and outpatient status(help to calculate inpatient and outpatient count)
df['Is_Inpatient'] = (df['Patient_Type'] == 'Inpatient').astype(int)
df['Is_Outpatient'] = (df['Patient_Type'] == 'Outpatient').astype(int)
df
```
